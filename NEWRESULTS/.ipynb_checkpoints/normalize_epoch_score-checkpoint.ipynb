```python
import pandas as pd
import numpy as np
```

```python
def give_normal_score(filename):
    data = pd.read_csv(filename, sep=":", header=None)
    data.columns = ["word", "score"]
    data.set_index('word',inplace=True)
    least_score = data['score'].min()
    data['n_score'] = np.where(data['score'].between(least_score,abs(least_score)), 0, data['score'])
    return data
```

FOR NOVICE

```python
p = give_normal_score('cosine0/cos_badminton_novice0.txt')
p1 = give_normal_score('cosine1/cos_badminton_novice1.txt')
p2 = give_normal_score('cosine2/cos_badminton_novice2.txt')
p3 = give_normal_score('cosine3/cos_badminton_novice3.txt')
p4 = give_normal_score('cosine4/cos_badminton_novice4.txt')
p5 = pd.DataFrame(index=p.index)
p5['epoch0'] = p['n_score']
p5['epoch1'] = p1['n_score']
p5['epoch2'] = p2['n_score']
p5['epoch3'] = p3['n_score']
p5['epoch4'] = p4['n_score']
print(p5)
```
